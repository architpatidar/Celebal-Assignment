```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.drop(columns=['Cabin'],inplace=True)
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'].fillna('S', inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='Survived', data = df)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Pclass', data = df)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
sns.histplot(x='Age', data = df, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='Sex', data = df)
plt.title('Sex Distribution')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data = df)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data = df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.histplot(x='Age', hue='Survived', data = df, bins=40)
plt.title('Survival by Age Group')
plt.show()
```

```python
sns.catplot(x='Sex', hue='Survived', col='Pclass', kind='count', data = df)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Embarked', hue='Survived', data = df, palette='Spectral')
plt.title('Survival Based on Embarkation Point')
plt.xlabel('Embarkation Point')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df, x='Fare', hue='Survived', bins=20, kde=True, palette='viridis')
plt.title('Fare Distribution by Survival')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
df['family_size'] = df['SibSp'] + df['Parch']
plt.figure(figsize=(10, 6))
sns.countplot(x='family_size', hue='Survived', data = df, palette='cubehelix')
plt.title('Survival Based on Family Size')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
bins = [0, 12, 18, 40, 60, 80]
labels = ['Child', 'Teenager', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)

plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', hue='Survived', data = df, palette='viridis')
plt.title('Survival Based on Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(12, 8))
numeric_features = df.select_dtypes(include=np.number)
sns.heatmap(numeric_features.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```
